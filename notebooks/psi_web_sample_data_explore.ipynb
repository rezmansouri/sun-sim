## kpo_mas

```python
v0, v0d0, v0d1 = read_hdf(
    "/Users/reza/Career/DMLab/SURROGATE/Data/psi_web_sample/cr1625/kpo_mas_mas_std_0101/br_r0.hdf",
    True,
)
v, vd0, vd1, vd2 = read_hdf(
    "/Users/reza/Career/DMLab/SURROGATE/Data/psi_web_sample/cr1625/kpo_mas_mas_std_0101/br002.hdf",
    False,
)

print(v0.shape, v.shape)

vd2 = np.expand_dims(vd2, axis=0)
vd2 = np.expand_dims(vd2, axis=0)

v0 = v0.flatten()

plt.figure(figsize=(20, 5))

plt.subplot(1, 3, 1)
plt.title("br_r0")
plt.hist(v0, bins=50)
plt.xlabel("Value")
plt.ylabel("Frequency")


plt.subplot(1, 3, 2)
plt.title("br002: original")
plt.hist(v.flatten(), bins=50)
plt.xlabel("Value")
plt.ylabel("Frequency")

vv = v / vd2

# plt.subplot(1, 4, 3)
# plt.title("br002: b = b / distance")
# plt.hist(vv.flatten(), bins=50)
# plt.xlabel("Value")
# plt.ylabel("Frequency")

vvv = v * (vd2 ** 2)

plt.subplot(1, 3, 3)
plt.title("br002: b = b * distance ^ 2")
plt.hist(vvv.flatten(), bins=50)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()


plt.figure(figsize=(20, 5))
for i in range(20, 140, 20):
    vvvv = v[:, :, i]
    plt.subplot(1, 6, (i-20) // 20 + 1)
    plt.title(f"br002: original at k={i}")
    plt.hist(vvvv.flatten(), bins=50)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(20, 5))
for i in range(20, 140, 20):
    vvvv = v[:, :, i] / vd2[:, :, i]
    plt.subplot(1, 6, (i-20) // 20 + 1)
    plt.title(f"br002: b/distance at k={i}")
    plt.hist(vvvv.flatten(), bins=50)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(20, 5))
for i in range(20, 140, 20):
    vvvv = v[:, :, i] * (vd2[:, :, i] ** 2)
    plt.subplot(1, 6, (i-20) // 20 + 1)
    plt.title(f"br002: b*distance^2 at k={i}")
    plt.hist(vvvv.flatten(), bins=50)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
plt.show()
```
