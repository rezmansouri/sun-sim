```python
import os
import numpy as np
from PIL import Image
from io import BytesIO
from pyhdf.SD import SD, SDC
import imageio.v2 as imageio
from tqdm import tqdm, trange
import matplotlib.pyplot as plt
```

```python
def read_hdf(path):
    f = SD(path, SDC.READ)
    arr = f.select("Data-Set-2").get()
    return arr
```

```python
arr = read_hdf(
    "/Users/reza/Career/DMLab/SURROGATE/Data/psi_web_sample/test/cr2233/hmi_masp_mas_std_0201/br002.hdf"
)
arr = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
slice = arr[:, :, 100].transpose(1, 0)
plt.figure(figsize=(8, 6))
im = plt.imshow(slice, cmap="jet_r")
plt.colorbar(im)
plt.show()
```

```python
frames = []
for i in trange(arr.shape[2]):
    slice = arr[:, :, i].transpose(1, 0)
    plt.figure(figsize=(8, 6))
    im = plt.imshow(slice, cmap="jet_r")
    plt.title(f'slice {i+1}')
    plt.colorbar(im)
    buf = BytesIO()
    plt.savefig(buf, format="png")
    frames.append(imageio.imread(buf))
    plt.close()
```

```python
output_filename = f"b.gif"
fps = 10  # Adjust frames per second as needed

# Create video directly from frames in memory
with imageio.get_writer(output_filename, fps=fps, loop=0) as writer:
    for frame in frames:
        writer.append_data(frame)
```
