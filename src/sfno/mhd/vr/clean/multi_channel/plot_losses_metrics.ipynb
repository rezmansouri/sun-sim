```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
result_path = "/Users/reza/Career/DMLab/SURROGATE/results/sfno/exp_35"
files = os.listdir(result_path)
best_epoch_name = [file for file in files if file.startswith("best_epoch")][0]
best_epoch = int(
    best_epoch_name[best_epoch_name.index("-") + 1 : best_epoch_name.index(".")]
)
print("best epoch:", best_epoch)
```

```python
train_losses, val_losses = np.load(
    os.path.join(result_path, "train_losses.npy")
), np.load(os.path.join(result_path, "val_losses.npy"))
train_rmse, val_rmse = np.load(os.path.join(result_path, "train_rmse.npy")), np.load(
    os.path.join(result_path, "val_rmse.npy")
)
train_nnse, val_nnse = np.load(os.path.join(result_path, "train_nnse.npy")), np.load(
    os.path.join(result_path, "val_nnse.npy")
)
train_msssim, val_msssim = np.load(
    os.path.join(result_path, "train_msssim.npy")
), np.load(os.path.join(result_path, "val_msssim.npy"))
train_acc, val_acc = np.load(os.path.join(result_path, "train_acc.npy")), np.load(
    os.path.join(result_path, "val_acc.npy")
)
train_psnr, val_psnr = np.load(os.path.join(result_path, "train_psnr.npy")), np.load(
    os.path.join(result_path, "val_psnr.npy")
)
```

```python
print('best val_losses:', val_losses[best_epoch-1])
print('best val_rmse:', val_rmse[best_epoch-1])
print('best val_nnse:', val_nnse[best_epoch-1])
print('best val_msssim:', val_msssim[best_epoch-1])
print('best val_acc:', val_acc[best_epoch-1])
print('best val_psnr:', val_psnr[best_epoch-1])
```

```python
# Create the plot
plt.figure(figsize=(10, 15))
epochs = np.arange(1, len(train_losses) + 1)

plt.subplot(3, 2, 1)
plt.plot(epochs, train_losses, label="Training Loss")
plt.plot(epochs, val_losses, label="Validation Loss")
# Add labels, title, and legend
plt.title("Loss")
plt.legend()
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(epochs, train_rmse, label="Training rmse")
plt.plot(epochs, val_rmse, label="Validation rmse")
# Add labels, title, and legend
plt.title("rmse")
plt.legend()
plt.grid(True)

plt.subplot(3, 2, 3)
plt.plot(epochs, train_nnse, label="Training NNSE")
plt.plot(epochs, val_nnse, label="Validation NNSE")
# Add labels, title, and legend
plt.title("NNSE")
plt.legend()
plt.grid(True)

plt.subplot(3, 2, 4)
plt.plot(epochs, train_msssim, label="Training MSSSIM")
plt.plot(epochs, val_msssim, label="Validation MSSSIM")
# Add labels, title, and legend
plt.title("MSSSIM")
plt.legend()
plt.grid(True)

plt.subplot(3, 2, 5)
plt.plot(epochs, train_acc, label="Training acc")
plt.plot(epochs, val_acc, label="Validation acc")
# Add labels, title, and legend
plt.xlabel("Epochs")
plt.title("acc")
plt.legend()
plt.grid(True)


plt.subplot(3, 2, 6)
plt.plot(epochs, train_psnr, label="Training psnr")
plt.plot(epochs, val_psnr, label="Validation psnr")
# Add labels, title, and legend
plt.xlabel("Epochs")
plt.title("psnr")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
